```r
decathlon.active <- decathlon2[1:23, 1:10]

# 手动标准化
# std.fun <- function(col) {
#     mean.col <- mean(col)
#     std.col <- sd(col)
#     return((col-mean.col)/std.col)
# }
# decathlon.active.std <- apply(decathlon.active, 2, std.fun)

decathlon.active.std  <- scale(decathlon.active)
head(decathlon.active.std)
```

```r
S <- cov(decathlon.active.std)
S
```

```r
eig.data <- eigen(S)
eig.data$values
```

```r
dimN <- ncol(decathlon.active.std)
rownames <- sapply(1:dimN, function(x){paste("Dim.", sep = "", x)})
eig.val <- data.frame(eigenvalues = eig.data$values,
                      row.names = rownames)
head(eig.val)
```

```r
# 10个维度所有方差（当然这里也是特征值）的和就是10，因为已经标准化过了
eig.val["variance.percent"] = eig.val["eigenvalues"]*dimN
eig.val["cumulative.variance.percent"] = cumsum(eig.val["variance.percent"])
eig.val
```

```r
barplot(eig.val$variance.percent,)
```

```r
# ggplot2
library(ggplot2)
# Minimal theme + blue fill color
# reference:http://www.sthda.com/english/wiki/ggplot2-barplots-quick-start-guide-r-software-and-data-visualization
p<-ggplot(data=eig.val, aes(x=rownames(eig.val), y=variance.percent)) +
  geom_bar(stat="identity", fill="steelblue")+
  scale_x_discrete(limits=rownames(eig.val))+
  geom_line(aes(x=(1:10), y=variance.percent), colour = "black")+
  geom_point(aes(x=(1:10), y=variance.percent), colour = "black")+
  geom_text(aes(label=sprintf("%1.2f%%", variance.percent)), 
            vjust=1.6, color="black", size=3.5)+
  labs(title="Scree Plot",x="Dimensions", y="Percentage of explained variance")+
  theme_minimal()
p
```

```r
# 选取前五个维度
m <- 5
# 因子载荷阵
A <- eig.data$vectors[, 1:m]
A
```

```r
# 主成分得分
F <- decathlon.active.std %*% A
head(F)
```

```r
# 用于综合评价
# 权重W
W <- eig.val$eigenvalues[1:m]/sum(eig.val$eigenvalues[1:m])
W
```

```r
# 评价函数Z = FW
Z <- data.frame(score=F %*% W)
head(Z)
```

```r
Z.sorted <- Z[order(Z["score"], decreasing = TRUE), , drop = FALSE]
head(Z.sorted)
```
