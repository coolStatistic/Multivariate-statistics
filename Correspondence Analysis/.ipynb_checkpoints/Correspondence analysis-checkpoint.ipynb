## Correspondence analysis

参考:*Practical Guide to Principal	Component Methods in R*, by Alboukadel

### R packages

```r
library("FactoMineR")
library("factoextra")
library("gplots")
options(repr.plot.width=7, repr.plot.height=4.5)
```

### Data format

```r
data(housetasks)
head(housetasks)
```

### Graph	of	contingency	tables	and	chi-square	test

```r
dt <- as.table(as.matrix(housetasks))
balloonplot(t(dt), main="housetasks", xlab = "",
           ylab = "", label = FALSE, show.margins = FALSE)
```

```r
# 卡方检验（行列是否相关）
chisq <- chisq.test(housetasks)
chisq
```

结果表明数据的行变量和列变量显著相关。

### R	code	to	compute	CA

```r
res.ca <- CA(housetasks, graph = TRUE)
```

```r
print(res.ca)
```

### Biplot options

#### Symmetric	biplot
>With symmetric plot, the inter-distance between rows and cols can't
be interpreted, Only general statements can be made about the pattern.

```r
options(repr.plot.width=5, repr.plot.height=4.5)
fviz_ca_biplot(res.ca, repel = TRUE)
```

>Note that, in order to interpret the distance between column points
and row points, the simplest way is to make an asymmetric plot. This means
that, the column profiles must be presented in row space or vice-versa. 

#### Asymmetric biplot

```r
fviz_ca_biplot(res.ca, map="rowprincipal", arrow=c(TRUE, TRUE),
              repel = TRUE)
```

>If the angle between two arrows is acute, then there is a strong 
association between the corresponding row and column.

#### Contribution biplot

```r
fviz_ca_biplot(res.ca, map="colgreen", arrow=c(TRUE, FALSE),
              repel = TRUE)
```

>The closer an arrow is(in terms of angular distance) to an axis the
greater is the contribution of the row category on that axis relative to the other 
axis. If the arrow is halfway between the two, its row category contributes	to the two axes to the same	extent.

### Filtering results

```r
# Visualize the rows with  cos2 >= 0.8
fviz_ca_row(res.ca, select.row = list(cos2=0.8))
```

```r
# Top 5	active rows and	5 suppl rows with the highest cos2
fviz_ca_row(res.ca, select.row = list(cos2=5), repel = TRUE)
```

```r
# Select by	names
name <-	list(name =	c("employment",	"fear",	"future"))
fviz_ca_row(res.ca,	select.row = name)
```
